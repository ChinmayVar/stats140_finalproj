```python
import pandas as pd
```

## Data

```python
all_stars_df = pd.read_csv('all-star-selections.csv')
players_df = pd.read_csv('playertotals.csv')
```

```python
players_df[players_df['player'] == 'Chris Gatling']
```

```python
all_stars_df.columns
```

```python
all_stars_df = all_stars_df.set_index(['season', 'player'])
all_stars_df
```

```python
players_df.columns
```

There is issue with players having team 'TOT', we will remove these rows

```python
players_df = players_df[players_df['tm'] != 'TOT']
```

```python
players_df = players_df[['season', 'player', 'player_id', 'g', 'gs', 'mp', 'fg', 'fga', 'x3p', 'x3pa', 'x2p', 'x2pa', 'ft', 'fta', 'orb', 'drb', 'ast', 'stl', 'blk', 'tov', 'pf', 'pts']]
```

Some features are only available starting with the 1982 season (we will only consider these). Also the 2025 all-star selection data is not available in the dataset (as of 3/5/2025) so we will exclude this as well.

```python
players_df = players_df[players_df['season'].between(1982, 2024)]
```

Finally we add a column for if a player was an all-star in a given season

```python
players_df = players_df.groupby(['season', 'player', 'player_id']).sum()
players_df['all_star'] = players_df.index.isin(all_stars_df.index)
players_df
```

```python
players_df['all_star'].value_counts()
```

```python
players_df.to_csv('all-star-dataset.csv')
```
